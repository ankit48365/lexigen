1. Tokenization ~ Tokenize the input text. This means we'll split the text into individual words (or tokens) and create a vocabulary of unique words.


```python
def tokenize(text):
    tokens = text.lower().split()
    return tokens

def build_vocab(tokens):
    vocab = list(set(tokens)) # set removes duplicate words and list converts it back to a list
    return vocab

text = "Hello, this is a simple text generator example. This is a simple example."
tokens = tokenize(text)
vocab = build_vocab(tokens)
print("Tokens:", tokens)
print("Vocabulary:", vocab)
```

Encoding ~ Now encode the tokens into numerical vectors. This involves creating one-hot vectors for each word in the vocabulary.

```python
import numpy as np

def one_hot_encode(tokens, vocab):
    vocab_size = len(vocab)
    print("Vocabulary Size:", vocab_size)
    token_to_idx = {token: idx for idx, token in enumerate(vocab)} # token_to_idx = {'a': 0, 'hello,': 1, 'example.': 2, 'generator': 3, 'simple': 4, 'text': 5, 'this': 6, 'is': 7}
    encoded_tokens = np.zeros((len(tokens), vocab_size))
    for i, token in enumerate(tokens):
        encoded_tokens[i, token_to_idx[token]] = 1
    return encoded_tokens

encoded_tokens = one_hot_encode(tokens, vocab)
print("Encoded Tokens:\n", encoded_tokens)
```
